```python
import pandas, re
import matplotlib.pyplot as plt
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
import speech_recognition as sr
from gtts import gTTS
from keras.models import load_model
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM
import os
from sklearn.model_selection import train_test_split
```

```python
## Verinin dosyadan okunması, parçalanması ve eğitim için hazırlanması
veri = pandas.read_csv("_data.csv", header=0, delimiter="\t", quoting=3)
veri = veri[['Review', 'Sentiment']]

turkceKarakterler = "a|b|c|ç|d|e|f|g|ğ|h|ı|i|j|k|l|m|n|o|ö|p|r|s|ş|t|u|ü|v|y|z|0-9"

veri['Review'] = veri['Review'].apply(lambda x: x.lower())# tüm kelimeleri küçük harfe dönüştürüyor
veri['Review'] = veri['Review'].apply((lambda x: re.sub('[^'+turkceKarakterler+'\s]','',x)))
#cümleyi parçalara ayırıyor
tokenizer = Tokenizer(split=' ',num_words=25000)# en sık  geçen kelimeye odaklanır
tokenizer.fit_on_texts(veri['Review'].values)
#tüm cümeleleri eşit boyuta getiriyor
X = tokenizer.texts_to_sequences(veri['Review'].values)
X = pad_sequences(X,maxlen=400)
Y = pandas.get_dummies(veri['Sentiment']).values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
#eğitim %80 test %20
## Model oluşturulması, eğitimi ve kaydedilmesi
embed_dim = 128
lstm_out = 128
model = Sequential()
model.add(Embedding(25000, embed_dim,input_length = X.shape[1]))
model.add(LSTM(lstm_out, dropout=0.2))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
# Oluşturulan model train verileri ile eğitilir. Yapay Sinir Ağı eğitilmeye başlar.
# epoch: İterasyon sayısı
history=model.fit(X_train, Y_train, epochs= 3, batch_size=32, verbose = 2)
# Train verileri ile model eğitildikten sonra test dataları ile doğruluk oranlarına bakılır.
score = model.evaluate(X_test, Y_test, verbose = 2)
print("score: %.2f" % (score[1]))

model.save("model.h5")


```

```python
#modelin yapılandırmasını kontrol etmek için kullanırız
model.summary()
```

```python
plt.figure(figsize = (15,10))

plt.plot(history.history['accuracy'])

plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test',], loc='upper left')
plt.show()
```
